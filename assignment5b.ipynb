#### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
# We can use LookupError exception class to handle both KeyError and IndexError exception classes.
```

```python
# A KeyError Exception occurs when we try to access a key that is not present.

Ansh_info = {'name': 'Ansh Rao',
                'age': 21,
                'language': 'Python'}
user_input = input('What do you want to learn about Ansh==> ')

try:
    print(f'{user_input} is {Ansh_info[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

```python
# IndexError is raised when a sequence reference is out of range.

x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```
